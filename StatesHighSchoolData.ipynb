```python
import pandas as pd
import geojson
import json
import geopandas as gpd
import numpy as np
from pandas import json_normalize
```

```python
with open('cb_2020_us_state_5m.geojson') as geojson_data:
    geometry_data = json.load(geojson_data)
```

```python
df_geojson = pd.DataFrame.from_dict(json_normalize(geometry_data['features']), orient='columns')
```

```python
df_geojson
```

```python
df_high_schools = pd.read_csv('ACSDT1Y2019.B14001_data_with_overlays_2021-08-04T200120.csv')
df_high_schools
```

```python
merged_df = pd.merge(df_high_schools, df_geojson, left_on='NAME', right_on='properties.NAME')
```

```python
merged_df
```

```python
merged_df.rename(columns= {'B14001_001E':'properties.Total_Num_of_students'}, inplace = True)
```

```python




merged_df.to_file("hs_data.geojson", driver="GeoJSON")
```
